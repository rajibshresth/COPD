```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read all the csv Data
data = pd.read_csv('./Encoded_Data.csv')
```

```python
# Show the first few rows
data.head()
```

```python
# Distribution of the Target Variable (COPD_Diagnosis)
plt.figure(figsize=(6, 4))
sns.countplot(x=data['COPD_Diagnosis'])
plt.title("Distribution of COPD Diagnosis")
plt.xlabel("COPD Diagnosis (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of Numerical Features
plt.figure(figsize=(16, 10))
for i, column in enumerate(['Age', 'BMI', 'Air_Pollution_Level']):
    plt.subplot(2, 2, i+1)
    sns.histplot(data[column], kde=True, bins=20)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Distribution of Categorical Features
plt.figure(figsize=(18, 10))
for i, column in enumerate(['Gender_Male', 'Smoking_Status_Former', 'Smoking_Status_Never']):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=data[column])
    plt.title(f"Distribution of {column}")
    plt.xticks(rotation=45)
    plt.xlabel(column)
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis using the encoded data
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
