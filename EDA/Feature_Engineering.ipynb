```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Read all the csv Data
data = pd.read_csv('./Encoded_Data.csv')
```

```python
# Show the first few rows
data.head()
```

```python
# Create Interaction Terms
data['Gender_Smoking'] = data['Gender_Male'] * data['Smoking_Status_Former'] * data['Smoking_Status_Never']
```

```python
# Log Transformation for Air_Pollution_Level and BMI
data['Air_Pollution_Level_log'] = data['Air_Pollution_Level'].apply(lambda x: np.log1p(x))  # log1p to handle zeroes
data['BMI_log'] = data['BMI'].apply(lambda x: np.log1p(x))

# Age Binning
bins = [0, 18, 40, 60, 100]
labels = ['Young', 'Adult', 'Middle_Aged', 'Elderly']
data['Age_Bin'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
data_encoded = pd.get_dummies(data, columns=['Age_Bin'], drop_first=True)
```

```python
# Save the data after feature Engineering
data_encoded.to_csv('../Prediction/COPD_Data_Engineered.csv', index=False, encoding='utf-8')
```
